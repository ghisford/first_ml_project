### Selecting columns using loc

> the loc attribute of a dataframe can be used to slice that dataframe. the general notation is dataframe.loc[row/index_name/value or row slice, column_name/ column slice]


> although it is mostly label based, it can also be used with a boolean array as we shall see later.





```python
# slicing the rm series- empty row slice should be noted
df.loc[:, 'rm']
```

```python
# the row indices are inclusive- both of them.
df.loc[2:5, ["rm","lstat"]]
```

```python
# one is a series the other is a dataframe. I hope you can see why
type(df.loc[:, 'rm'])
type(df.loc[:, ["rm"]])
```

```python
# understanding the use of the colon is for rows only
# slicing rows from index 5-10
df.loc[5:10,['rm','lstat']]
```

```python
# including the step ...
df.loc[0:10:2, ['rm', 'lstat']]
```

```python
# getting a specific row
df.loc[2,['rm','lstat']]
```

```python
# getting all the columns and row using the :
df.loc[:, :]
```

```python
# loc allows you to pick a single row with all its columns
df.loc[3]
```

```python
# it also enables you to pick a range of columns without explicitly naming them all
df.loc[3, 'crim':'rm']
```

## What do you understand by filtering ?
It is picking data that satisfies certain conditions

```python
import pandas as pd
filename1 = '/content/drive/MyDrive/mortgages.csv'
dff = pd.read_csv(filename1)
dff
```

```python
# lets get some basic info about our data

dff.head()
```

```python
# getting the number of rows and columns
dff.shape
```

```python
# get info about the datatypes
dff.info()
```

```python
# seeing where the null values are or arent
dff.isnull()
```

```python
# counting all the non-null values per column
dff.isnull().count()
```

```python
# counting the null values using sum()
dff.isnull().sum()
```

```python
# lets print the columns
dff.columns
```

```python
# the value_counts() method returns the number of occurences of the unique values
# in a series- for categorical data
dff['Mortgage Name'].value_counts()
```

```python
# We can also get this for the interest rate series
dff["Interest Rate"].value_counts()
```

```python
# Question: How many mortgages have a 30yr term?
```

```python
# How do we create filters?
# we can use comparison operators e.g <,>,!=, ==

# we can get the 30 year mortgages using the ==

dff["Mortgage Name"] == '30 Year'
# this filter returns a boolean series/ array based on the evaluation of the condition
```

```python
# storing this filter in a variable for future use...
mortgage_name_filter = dff["Mortgage Name"] == "30 Year"
```

```python
# filtering our dataframe...
# notation is dataframe[filter] and this will return a filtered dataframe eg
dff[mortgage_name_filter]
```

```python
# storing it in a new dataframe
dff_filtered =  dff[mortgage_name_filter]
```

```python
# checking the value_counts for the mortgage name filters
dff_filtered["Mortgage Name"].value_counts()
# yay! we successfully eliminated anything that wasnt '30 year'
```

```python
# lets create a filter for interest rate without looking
# creating the filter
Interest_filter = dff["Interest Rate"] == 0.05
```

```python
# creating a new dataframe using that filter
dff_inter_filtered= dff[Interest_filter]
# testing the unique Interest rates inside that dataframe
dff_inter_filtered["Interest Rate"].value_counts()
```

```python
# Fancy: filter using the loc attribute
# notatin dataframe.loc[filter, columns]
dff.loc[Interest_filter, :].head()
# this allows me to filter and slice at the same time in an elegant way
```

## **Combining Filters**
Using bitwise operators eg &, | to combine filters

```python
# combine the above 2 filters...
# get data where Mortgage is 30 Year and Interest rate is 0.05

# creating a new dataframe that uses the 2 filters

df_combined_filter = dff[mortgage_name_filter&Interest_filter]
df_combined_filter.head()
```

```python
# lets do a value count and see whether anything changed from last time
df_combined_filter["Interest Rate"].value_counts()
```

```python
df_combined_filter["Mortgage Name"].value_counts()
```

```python
# Question: show data where the mortgage is 15 years or interest rate is less than
# 0.04
mortgage_filter = dff["Mortgage Name"] == "15 Year"
interest_filter2 = dff["Interest Rate"] == 0.04

dff_combined = dff[mortgage_filter|interest_filter2]

dff_combined.head()
```
